```python
#mean
import numpy as np
scores=[10,20,100,200]
print(np.mean(scores))
```

```python
#median
import numpy as np
scores=[10,40,70,100]
median_scores=np.median(scores)
print(median_scores)
```

```python
#MODE
from statistics import mode
votes=[4,7,8,5,9,9,8,8,8]
most_common=mode(votes)
print(most_common)
```

```python
#variance and standrad deviation
import numpy as np
tight_data=[40,50,60]
spread_data=[30,120,50]
print("tight_data:",np.std(tight_data))
print("spread_data:",np.std(spread_data))
```

```python
import random
samples=[random.randint(1,5) for _ in range(8)]
print("sample rolls:",samples)
print("sample mean:",sum(samples)/len(samples))
```
